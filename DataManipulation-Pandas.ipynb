### Data Manipulation with Pandas

```python
#Import Package
```

```python
import pandas as pd
import numpy as np
```

```python
#Create DataFrame
```

```python
df1 = pd.DataFrame({
    "name": ["Anna", "Dane", "David", "Kevin", "Joe", "Rose"],
    "weight": [45.2,50.5,62.1,64.0,59.0,np.nan],
    "height": [155,160,162,170,167,159],
    "birth_year": [2000,1997,1996,1987,1998,2003],
    "gender": ["female", "male", "male", "male", "male", "female"],
    "city": ["New York", "Chicago", "Austin", "New York", np.nan, "New York"],
    })
df1
```

```python
#Create another Pandas DataFrame
```

```python
df2 = pd.DataFrame({
    "nick_name": ["Anna", "David", "Dane", "Rose"],
    "full_name": ["Anna Levis", "David John", "Dane  Sebastian", "Rose Kimberly"]
    })
df2
```

```python
#Merge DataFrame (key = name and nick_name)
```

```python
df = df1.merge(df2, how = "left", left_on = "name", right_on = "nick_name")
df
```

```python
#Drop nick_name column
```

```python
df
```

```python
#Fill null values with their original name column
```

```python
df.full_name.fillna(df.name, inplace = True)
df
```

```python
#Create a new BMI column calculated based on the following formula
```

```python
#Create a new function to calculated formula of bmi
def custom_bmi(weight, height) :
    return weight/(height/100)**2

#apply the function to DataFrame using lambda
df['BMI'] = df.apply(lambda x: custom_bmi(x.weight, x.height), axis = 1).fillna(0)
df
```

```python
#Create a new 'bmi_class' column following the definition
```

```python
#Create a new function to calculated bmi_class
def bmi_class(bmi) :
  if bmi >= 40 :
    return "Obesity class 3"
  elif bmi >= 35 :
    return "Obesity class 2"
  elif bmi >= 30 :
    return "Obesity class 1"
  elif bmi >= 25 :
    return "Overweight"
  elif bmi >= 18.5 :
    return "Normal weight"
  else :
    return "Underweight"

#apply the function to DataFrame
df['Class'] = df['BMI'].apply(bmi_class)
df
```
